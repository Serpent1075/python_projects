```python
import pandas as pd
import numpy as np
from pathlib import Path

data_path = Path('../../data_analysis_lect/datasets/Covid19-US/us_confirmed.csv')
df = pd.read_csv(data_path)
df.info()
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# head
df.head()
```

```python
# datetime index를 가진 테이블의 시간 슬라이싱
df = df.set_index('Date').sort_values('Date')
df['2020-01':'2020-02']
```

```python
df.index.unique()
```

```python
df = df.reset_index()
df.index.unique()
```

```python
# 문자열 index를 가진 데이터테이블
df = df.reset_index().set_index('Province/State').sort_index() #reset_index : index 초기화 함
df.info()
```
